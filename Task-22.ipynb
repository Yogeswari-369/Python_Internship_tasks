```python
# 2.	Merge two pdf files using python script


from PyPDF2 import PdfFileReader, PdfFileMerger
pdf_file1 = PdfFileReader("Task-19.pdf")
pdf_file2 = PdfFileReader("Task-20.pdf")
output = PdfFileMerger()
output.append(pdf_file1)
output.append(pdf_file2)
with open("merged.pdf", "wb") as output_stream:
    output.write(output_stream)
print(" merging completed..!")
```

```python
# 3.	Scrape a website and store the data into DB.

page=urllib.request.urlopen("https://docs.python.org/3/library/random.html")
soup=bs4(page)
names=soup.body.findAll('dt')

func_names= re.findall('id="random.\w+',str(names))
func_names=[item[4:] for item in func_names]
des=soup.body.findAll('dd')
func_usage=[]

for item in des:
  item=item.text
  item=item.replace("\n"," ")
  func_usage.append(item)

print('list of fun names:',func_names[:5])
print('\n func des:',func_usage[0])
print('\n number of items in func names:',len(func_names))
print('\n no. of items in func des:', len(func_usage))



```

```python
func des: Initialize the random number generator.
If a is omitted or None, the current system time is used.  If randomness sources are provided by the operating system,
they are used instead of the system time (see the os.urandom() function for details on availability). 
If a is an int, it is used directly. With version 2 (the default), a str, bytes, or bytearray object gets converted to an int and all of its bits are used.
With version 1 (provided for reproducing random sequences from older versions of Python), the algorithm for str and bytes generates a narrower range of seeds.  
Changed in version 3.2: Moved to the version 2 scheme which uses all of the bits in a string seed.   
Deprecated since version 3.9: In the future, the seed must be one of the following types: NoneType, int, float, str, bytes, or bytearray.  
```
